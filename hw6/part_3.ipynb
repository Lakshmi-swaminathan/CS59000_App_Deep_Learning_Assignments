```python
from google.colab import drive
import os

# Mount Google Drive
drive.mount('/content/drive')

# Set paths for test dataset
test_real_path = '/content/drive/MyDrive/cifake_dataset/test/REAL'
test_fake_path = '/content/drive/MyDrive/cifake_dataset/test/FAKE'
```

```python
from transformers import pipeline

# Initialize the SDXL detector pipeline
pipe = pipeline("image-classification", model="Organika/sdxl-detector")
```

```python
from PIL import Image
import numpy as np

def calculate_accuracy(image_folder, target_label):
    correct_count = 0
    total_images = 0

    # Loop through each image in the folder
    for image_name in os.listdir(image_folder):
        image_path = os.path.join(image_folder, image_name)

        try:
            # Load the image
            image = Image.open(image_path).convert("RGB")

            # Run the image through the detector
            prediction = pipe(image)

            # Check if the prediction matches the target label
            if prediction[0]['label'].lower() == target_label:
                correct_count += 1

            total_images += 1

        except Exception as e:
            print(f"Error processing {image_name}: {e}")

    # Calculate accuracy
    accuracy = (correct_count / total_images) * 100 if total_images > 0 else 0
    return accuracy
```

```python
# Calculate accuracy for the REAL images (expected label: "human")
real_accuracy = calculate_accuracy(test_real_path, "human")
print(f"Detection accuracy for REAL images: {real_accuracy:.2f}%")

# Calculate accuracy for the FAKE images (expected label: "artificial")
fake_accuracy = calculate_accuracy(test_fake_path, "artificial")
print(f"Detection accuracy for FAKE images: {fake_accuracy:.2f}%")
```
