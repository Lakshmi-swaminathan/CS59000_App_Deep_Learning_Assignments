```python
import moviepy.editor as mp
import gtts

# List of images and their respective durations in seconds
images = [
    ("./image1.jpeg", 13),
    ("./image2.jpeg", 14),
    ("./image3.jpeg", 13)
]

# Corresponding voiceover text for each image
voiceover_texts = [
    "A sleek black cat sits on a window sill, its eyes fixed on the vibrant garden below. Curiosity sparks in its emerald eyes as it watches butterflies flit between colorful blooms.",
    "A golden retriever leaps through a field of wildflowers, its tail wagging with pure joy. The sun shines brightly, casting playful shadows on its golden fur as it chases a frisbee.",
    "A lone cyclist pedals through a serene forest, surrounded by towering trees and lush greenery. The gentle rustling of leaves and the chirping of birds create a peaceful ambiance."
]

# Generate voiceover using gTTS
voiceover_clips = []
for i, text in enumerate(voiceover_texts):
    tts = gtts.gTTS(text, lang='en')
    tts.save(f"./voiceover_{i}.mp3")
    voiceover_clips.append(mp.AudioFileClip(f"./voiceover_{i}.mp3"))

# Create a video clip for each image
image_clips = []
for i, (image_path, duration) in enumerate(images):
    img_clip = mp.ImageClip(image_path).set_duration(duration)
    img_clip = img_clip.set_audio(voiceover_clips[i])  # Attach the audio to the image
    image_clips.append(img_clip)

# Concatenate all image clips into one video
final_clip = mp.concatenate_videoclips(image_clips)

# Combine audio from all clips for the final output
final_audio = mp.concatenate_audioclips(voiceover_clips)

# Set the combined audio to the final video
final_clip = final_clip.set_audio(final_audio)

# Write the final video to a file
output_file_path = "./Application_of_Deep_Learning_Course.mp4"
final_clip.write_videofile(output_file_path, fps=24, codec='libx264', audio_codec='aac')

output_file_path
```
