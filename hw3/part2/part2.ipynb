```python
from PIL import Image, ImageDraw, ImageFont
import moviepy.editor as mp
import textwrap


# Load the existing video
video_path = "./Application_of_Deep_Learning_Course.mp4"
video = mp.VideoFileClip(video_path)

# List of captions and their respective durations in seconds
captions = [
    ("A sleek black cat sits on a window sill, its eyes fixed on the vibrant garden below. Curiosity sparks in its emerald eyes as it watches butterflies flit between colorful blooms.", 0, 13),
    ("A golden retriever leaps through a field of wildflowers, its tail wagging with pure joy. The sun shines brightly, casting playful shadows on its golden fur as it chases a frisbee.", 13, 27),
    ("A lone cyclist pedals through a serene forest, surrounded by towering trees and lush greenery. The gentle rustling of leaves and the chirping of birds create a peaceful ambiance.", 27, 40)
]

# Function to create a text image using PIL and wrapping the text into multiple lines
def create_text_image(text, size=video.size, max_width=40, fontsize=100):
    font = ImageFont.load_default()  # Using the default PIL font
    image = Image.new('RGBA', size, (0, 0, 0, 0))
    draw = ImageDraw.Draw(image)

    # Wrap text into multiple lines based on max_width
    wrapped_text = textwrap.fill(text, width=max_width)

    # Calculate text size to center it
    text_size = draw.textsize(wrapped_text, font=font)
    text_x = (size[0] - text_size[0]) / 2
    text_y = size[1] - text_size[1] - 100  # Padding from the bottom

    # Draw background rectangle with padding
    padding = 20
    background_size = (text_size[0] + 2 * padding, text_size[1] + 2 * padding)
    background_x = (size[0] - background_size[0]) / 2
    background_y = text_y - padding
    draw.rectangle([background_x, background_y, background_x + background_size[0], background_y + background_size[1]], fill=(0, 0, 0, 180))

    # Draw text
    draw.text((text_x, text_y), wrapped_text, font=font, fill=(255, 255, 255, 255))

    return image

# Create text clips for captions
caption_clips = []
for text, start_time, end_time in captions:
    text_image = create_text_image(text,(600, 200), max_width=60)  # Use max_width to break lines
    text_image.save(f"/tmp/caption_{start_time}.png")
    txt_clip = mp.ImageClip(f"/tmp/caption_{start_time}.png")
    txt_clip = txt_clip.set_position('bottom').set_duration(end_time - start_time).set_start(start_time)
    caption_clips.append(txt_clip)

# Overlay text clips on the video
video_with_captions = mp.CompositeVideoClip([video] + caption_clips)

# Write the final video with captions to a file
output_file_path_with_captions = "./Application_of_Deep_Learning_Course_with_Captions.mp4"
video_with_captions.write_videofile(output_file_path_with_captions, fps=24, codec='libx264', audio_codec='aac')

output_file_path_with_captions
```
